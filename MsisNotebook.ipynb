# MSIS Plotting Notebook
This is a companion to the python implementation of the MSIS model. It allows you to plot maps of the MSIS output and line plots of the MSIS drivers (F-10.7 solar flux proxy, and AP
planetary index) versus date.

```python
#Import everything and set up for inline figures
%load_ext autoreload
%autoreload 2
%pylab inline
#if using DavitPy, then uncomment the following line and comment out 'import msispy'
#import models.msis as msispy
import msispy
import numpy, datetime #I know it's not nessecary to import numpy if pylab, but I prefer to be explicit, less confusion
import matplotlib.pyplot as pp
import matplotlib as mpl

#Set some more sensible default fontsizes (for my setup **yrmv**)
fs = 18 #font size
mpl.rcParams['font.size']=fs
```

```python
#Define a simpified vectorized interface to the MSIS model
def msis(dt,lat,lon,alt):
	from collections import OrderedDict
	import datetime
	"""
	Calls the MSIS model for a single time and
	an array of GEODETIC latitudes and longitudes and altitudes

		INPUTS
		------
			dt - a single datetime.datetime object
			lat - a [m x 1] numpy array of latitudes
			lon - a [m x 1] numpy array of longitudes
			alt - a [m x 1] numpy array of altitudes

		RETURNS
		-------
			species - a OrderedDictionary of [m x 1] numpy arrays
				The number density of various atmomic species, keys
				are HE, O, N2, O2, AR, H, N, AO (anomalous oxygen)
				and the mass density, in g / cm^3 (key is 'mass')
			t_exo - an [m x 1] numpy array
				Exospheric temperature 
			t_alt - an [m x 1] numpy array
				temperature at altitude
            drivers - a dictionary with the following keys
				datetime - the datetime to which the drivers correspond 
                f107: daily f10.7 flux for previous day
				f107a: 81 day average of f10.7 flux (centered on date)
				ap: magnetic index containing:
				(1) daily AP
				(2) 3 HR AP index for current time
				(3) 3 HR AP index for 3 hours before current time
				(4) 3 HR AP index for 6 hours before current time
				(5) 3 HR AP index for 9 hours before current time
				(6) Average of eight 3 hour AP indicies from 12 to 33 hrs prior to current time
				(7) Average of eight 3 hour AP indicies from 36 to 57 hrs prior to current time   
	"""     
	# # Get the F10.7 and AP values for the current year
	drivers = msispy.getF107Ap(dt)
	npts = len(lat.flatten())
	temp = numpy.empty((npts,1))
	temp.fill(numpy.nan)
	#Preallocate for speed
	species = OrderedDict((('HE',temp.copy()), ('O',temp.copy()), ('N2',temp.copy()),('O2',temp.copy()),\
	 ('AR',temp.copy()),('mass',temp.copy()),('H',temp.copy()),('N',temp.copy()),('AO',temp.copy())))
	t_exo = temp.copy()
	t_alt = temp.copy()

	#Parse out the date and time
	#(2 digit year + 3 digit day of year as an integer) (yes...really)
	ydstr = "%.2d%.3d" % (numpy.mod(dt.year,100),dt.timetuple().tm_yday)
	iyd = int(ydstr)
	#UT second of day
	sec = (dt-datetime.datetime.combine(dt.date(),datetime.time(0))).total_seconds()
	for p in range(npts):
		slt = sec/3600.+lon.flatten()[p]/15. #Lazy compute local solar time for now
		d,t = msispy.gtd7(iyd,sec,alt.flatten()[p],lat.flatten()[p],lon.flatten()[p],\
					slt,drivers['f107'],drivers['f107a'],drivers['ap'],48)
		for i,atom in enumerate(species):
			species[atom][p,0] = d[i]
		t_exo[p] = t[0]
		t_alt[p] = t[1]

	return species,t_exo,t_alt,drivers
```

# Daily MSIS Drivers For a Full Year
This will generate an overview plot of the AP and F10.7 for a full year of your choice. Useful if you want to find a particular activity level to look at MSIS output for. MSIS uses much more than just daily F10.7 and AP (see description of 'drivers' output variable in above function). It should be straightforward to modify the following code to display others of them if you want.

```python
year = 2005.
hours = [12] #Hour(s) for which we will get data for each day 
isleapyear=numpy.mod(year,4)==0
days = numpy.arange(1,366) if isleapyear else numpy.arange(1,365)

f107 = []
ap = []
times = []
for day in days:
    for hour in hours:
        times.append(datetime.datetime(2005,1,1,12,0,0)+datetime.timedelta(days=day-1,hours=hour))
for dt in times:
    drivers = msispy.getF107Ap(dt)
    f107.append(drivers['f107']) #F107
    ap.append(drivers['ap'][1]) #3 hour ap
    if dt.day==1:
        print "Reading drivers for month %d" % (dt.month)
fig = pp.figure(figsize=(12,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(times,f107,'b-')
ax1.set_ylabel('F10.7')
ax2.plot(times,ap,'r-')
ax2.set_ylabel('3 Hour AP')
fig.autofmt_xdate()
```

# Altitude Versus Latitude, Colored by Oxygen Number Density 
These pseudocolor plots replicate (not exactly) some found in David Hansen's presentation which is near the top of google for MSIS (strangely).
https://spaceweather.usu.edu/files/uploads/PDF/SWC_TR2010_001_View.pdf

```python
#We want a logarithmic color scale
from matplotlib.colors import LogNorm
from matplotlib import ticker
locator = ticker.LogLocator()
formatter = ticker.LogFormatter(10, labelOnlyBase=False)

f = pp.figure(figsize=(16,10))
a = pp.axes()

nlats=20
nalts=20
lats = numpy.linspace(-90.,90.,num=nlats)
alts = numpy.linspace(0,800,num=nalts)
latgrid,altgrid = numpy.meshgrid(lats,alts)
lon=-90. #Minnesota?
dt = datetime.datetime(2005,12,21,18,0,0) #Pick december 21, solar max, 18 UT (12 local time @ 90 West)
#Call MSIS with array inputs
species,t_exo,t_alt,drivers = msis(dt,latgrid.flatten(),numpy.ones_like(altgrid.flatten())*lon,altgrid.flatten())
Z = numpy.reshape(species['O'],latgrid.shape)
Z[Z<=0.] = numpy.nan
mappable = a.pcolormesh(latgrid,altgrid,Z,norm=LogNorm(vmin=1e6,vmax=1e12),shading='gouraud')
pp.colorbar(mappable,ticks=locator,format=formatter)
a.set_xlabel('Geodetic Latitude')
a.set_ylabel('Altitude(km)')
a.set_title('MSIS Oxygen Number Density for %s'%(dt.strftime('%c')))
```
